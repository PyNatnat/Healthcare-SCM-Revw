```python
import nltk
from nltk.corpus import stopwords
from nltk import pos_tag, word_tokenize
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import os
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
```

```python
from google.colab import drive

# Montar o Google Drive
drive.mount('/content/drive')

# Funções para processamento de textos
def extrair_substantivos(texto):
    palavras = word_tokenize(texto.lower())
    palavras_filtradas = [palavra for palavra, pos in pos_tag(palavras) if pos.startswith('NN')]
    return palavras_filtradas

def gerar_nuvem_de_palavras(texto, output_image):
    stop_words = set(stopwords.words('portuguese'))  # Usar stopwords em português

    # Filtra apenas substantivos
    palavras_filtradas = extrair_substantivos(texto)

    # Remove stopwords
    palavras_filtradas = [palavra for palavra in palavras_filtradas if palavra not in stop_words]

    # Gera a nuvem de palavras
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(palavras_filtradas))

    # Exibir a nuvem de palavras
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

    # Salvar a imagem da nuvem de palavras
    wordcloud.to_file(output_image)

def processar_artigos(pasta_artigos, output_image='nuvem_palavras.png'):
    texto_completo = ""

    # Itera sobre os arquivos na pasta
    for filename in os.listdir(pasta_artigos):
        if filename.endswith(".txt"):  # Supondo que os artigos estejam em arquivos .txt
            with open(os.path.join(pasta_artigos, filename), 'r', encoding='utf-8') as file:
                texto_completo += file.read() + " "

    # Gera a nuvem de palavras para o texto completo
    gerar_nuvem_de_palavras(texto_completo, output_image)

# Caminho para a pasta contendo os artigos no Google Drive
pasta_artigos = "/content/drive/MyDrive/srl"
processar_artigos(pasta_artigos)

```
